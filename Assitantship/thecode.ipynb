```python
# -------------------- IMPORTING PACKAGES --------------------
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# -------------------- READING EXCEL FILE --------------------
excel_file_path = "/Users/udoychowdhury/Documents/Assitantship/Conditionally Admitted Students.xlsx"

# -------------------- READING EXCEL SHEETS --------------------
programmajor = pd.read_excel(excel_file_path, sheet_name='Program to Major to College Cro') 
courseattribute = pd.read_excel(excel_file_path, sheet_name='Course Attributes by CRN Report') 
codecrosswalk = pd.read_excel(excel_file_path, sheet_name='Code Crosswalk') 
springregistration = pd.read_excel(excel_file_path, sheet_name='Spring 2023 Registered Students')
conditionaladmits = pd.read_excel(excel_file_path, sheet_name='Conditional Admits') 
fallregistration = pd.read_excel(excel_file_path, sheet_name='Fall 2022 Registration No Drops')
```

```python
# -------------------- DISPLAY DATAFRAMES --------------------
display(programmajor)
```

Exploratory Data Analysis

```python
# -------------------- Program to Major to College Cro --------------------
# === Understand Structure ===
print(programmajor.info())
print(programmajor.describe())
# === Identify Missing Values ===
print(programmajor.isnull().sum())

# === Descriptive Statistics ===
# Count of each program
print(programmajor['Program'].value_counts())

# === Bar Plot ===
# Bar plot for College
programmajor['College'].value_counts().plot(kind='bar')
plt.title('Number of Programs per College')
plt.xlabel('College')
plt.ylabel('Number of Programs')
plt.show()
```

```python
# -------------------- Course Attributes by CRN Report --------------------
# === Understand Structure ===
print(courseattribute.info())
print(courseattribute.describe())
# === Identify Missing Values ===
print(courseattribute.isnull().sum())

# === Descriptive Statistics ===
# Since this dataset primarily contains categorical data, we'll use value_counts() instead of describe()
print(courseattribute['Subject'].value_counts())

# === Count Plot ===
# Count plot for Course Subject
sns.countplot(x=courseattribute['Subject'])
plt.title('Distribution of Courses by Subject')
plt.xticks(rotation=90)
plt.show()

```

```python
# -------------------- Code Crosswalk --------------------
# === Understand Structure ===
print(codecrosswalk.info())
print(codecrosswalk.describe())
# === Identify Missing Values ===
print(codecrosswalk.isnull().sum())

# === Descriptive Statistics ===
# Count of each code's occurrence
print(codecrosswalk['Code'].value_counts())
```

```python
# -------------------- Spring 2023 Registered Students --------------------
# === Understand Structure ===
print(springregistration.info())
print(springregistration.describe())
# === Identify Missing Values ===
print(springregistration.isnull().sum())

# === Bar Plot ===
# Bar plot for Class Level
springregistration['Class Level'].value_counts().plot(kind='bar')
plt.title('Distribution of Class Levels')
plt.xlabel('Class Level')
plt.ylabel('Count')
plt.show()

# === Grouped Analysis ===
# Average Credits Enrolled by Residency
print(springregistration.groupby('Residency')['Total Credits Enrolled'].mean())

```

```python
# -------------------- Conditional Admits --------------------
# === Understand Structure ===
print(conditionaladmits.info())
print(conditionaladmits.describe())
# === Identify Missing Values ===
print(conditionaladmits.isnull().sum())

# === Count Plot ===
# Count plot for ETHNICITY
sns.countplot(x=conditionaladmits['ETHNICITY'])
plt.title('Ethnicity Distribution')
plt.xticks(rotation=45)
plt.show()

# === Grouped Analysis ===
# Average SAT score by First-Generation Indicator
print(conditionaladmits.groupby('FIRST_GEN_IND')['SAT_TOTAL'].mean())


```

```python
# -------------------- Fall 2022 Registration No Drops --------------------
# === Understand Structure ===
print(fallregistration.info())
print(fallregistration.describe())
# === Identify Missing Values ===
print(fallregistration.isnull().sum())

# === Histogram for GPA ===
fallregistration['Term_GPA'].hist(bins=20)
plt.title('Term GPA Distribution')
plt.xlabel('Term GPA')
plt.ylabel('Frequency')
plt.show()

# === Box plot for GPA ===
fallregistration.boxplot(column=['Total_Credit_Hours'])
plt.title('Total Credit Hours Distribution')
plt.ylabel('Credit Hours')
plt.show()

# === Grouped Analysis ===
# Average GPA by Gender
print(fallregistration.groupby('Gender')['Term_GPA'].mean())

# Average Credit Hours by Program
print(fallregistration.groupby('Program')['Total_Credit_Hours'].mean())

```
