#### INNER JOIN (or JOIN)
Display matching records from TWO tables! Choose to combine two tables using inner join. Write your question you are answering with your data query. <br>
<br>
**Research Question:**  Which players have their overall rating improved in the latest FIFA version compared to their oldest FIFA version appearence?
<br>
**What tables are we joining?**  Player Info with Player Info

```python
query = """SELECT 
    a.player_id,
    a.short_name,
    a.overall AS overall_oldest_version,
    b.overall AS overall_latest_version
FROM 
    player_info AS a
INNER JOIN player_info AS b ON a.player_id = b.player_id
WHERE 
    a.fifa_version = (SELECT MIN(fifa_version) FROM player_info) AND
    b.fifa_version = (SELECT MAX(fifa_version) FROM player_info) AND
    a.overall < b.overall;
"""

results = pd.read_sql(query, con)

display(results.head())
```

**LEFT JOIN**
<br>
Returns ALL records from the left table and matching records from the right table. Write your question you are answering with your data query. 
<br><br>
**Research Question:** What is the potential and current overall in the latest FIFA?
<br>
**What tables are we joining?** Player Info and Player Attributes

```python
query = """SELECT 
    pi.player_id,
    pi.short_name,
    pi.potential,
    pa.overall
FROM 
    player_info pi
LEFT JOIN player_attributes pa ON pi.player_id = pa.player_id AND pi.fifa_version = pa.fifa_version
WHERE 
    pi.fifa_version = (SELECT MAX(fifa_version) FROM player_info);
"""

results = pd.read_sql(query, con)

display(results.head())
```

**RIGHT JOIN**
<br>
Returns ALL records from the right table and matching records from the left table. Write your question you are answering with your data query. <br>
<br>
**Research Question:** What are a players attributes?
<br>
**What tables are we joining?** Player Info and Player Attributes

```python
query = """SELECT 
    pa.*,
    pi.short_name
FROM 
    player_info pi
RIGHT JOIN player_attributes pa ON pi.player_id = pa.player_id AND pi.fifa_version = pa.fifa_version
WHERE 
    pa.fifa_version = 24;
"""

results = pd.read_sql(query, con)

display(results.head())
```

**Question:** What are the club and national teams players have been part of and their jersey number?
<br>
**What tables are we joining?** Player Info and Player Info

```python
query = """
(SELECT 
    pi.short_name AS player_name,
    pi.club_name AS team_name,
    pi.club_jersey_number AS kit_number,
    'Club' AS team_type,
    pi.fifa_version
FROM 
    player_info pi
WHERE 
    pi.club_name IS NOT NULL)

UNION 

(SELECT 
    pi.short_name AS player_name,
    pi.nationality_name AS team_name,
    pi.nation_jersey_number AS kit_number,
    'National' AS team_type,
    pi.fifa_version
FROM 
    player_info pi
WHERE 
    pi.nationality_name IS NOT NULL)

ORDER BY player_name, team_type, fifa_version;
"""

results = pd.read_sql(query, con)

display(results.head())
```
